## Boring part

We only have the user inputted title and author. We can query the Open Library API
to search, based on those two things, to get the ISBN and a bunch of other info.

```python
import requests
def get_book(query):
    return requests.get('http://openlibrary.org/search.json', {'q': query}).json()['docs'][0]
```

```python
# from http://stackoverflow.com/a/2635066/907060

from threading import Thread
from queue import Queue
import math

def doWork():
    while not q.empty():
        i, query = q.get_nowait()
        try:
            resp = get_book(query)
        except IndexError:
            pass
        else:
            log_books.set_value(i, 'description', resp)
        q.task_done()

q = Queue()

for i, row in log_books.iterrows():
    title = row.book_title
    if not 'No Title Logged' == title:
        author = row.book_author
        if not (isinstance(author, float) and math.isnan(author)):
            title += ' ' + author
        q.put((i, title))


for i in range(20):
    t = Thread(target=doWork)
    t.daemon = True
    t.start()
```

```python
from pandas.io.json import json_normalize

descriptions = log_books['description']
def isbn(d):
    try:
        return d['isbn'][0]
    except Exception:
        return None

def author(d):
    try:
        return d['author_name'][0]
    except Exception:
        return None

def publish_year(d):
    try:
        return d['first_publish_year']
    except Exception:
        return None

def title(d):
    try:
        return d['title']
    except Exception:
        return None

log_books['isbn'] = descriptions.apply(isbn)
log_books['author_name'] = descriptions.apply(author)
log_books['publish_year'] = descriptions.apply(publish_year)
log_books['title'] = descriptions.apply(title)
```
